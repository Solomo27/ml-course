```python
graph = {
"A": [ "B" , "C"] ,
'B': [ "D", "E"] ,
'C': [ 'F' ] ,
'D': [] ,
'E': [ 'F' ] ,
'F': []
}

from collections import deque

def bfs ( graph , start , goal ) :
  queue = deque ([[ start ]])
  visited = set ()

  while queue :
    path = queue . popleft ()
    node = path [ -1]

    if node == goal :
      return path

    if node not in visited :
      visited . add ( node )

    for neighbor in graph [ node ]:
      new_path = list ( path )
      new_path . append ( neighbor )
      queue . append ( new_path )

  return None
print("BFS Path : ",bfs( graph , "A" , "F"))
```
