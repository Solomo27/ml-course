```python
graph = {
"A": [ "B" , "C"] ,
'B': [ "D", "E"] ,
'C': [ 'F' ] ,
'D': [] ,
'E': [ 'F' ] ,
'F': []
}


def bfs ( graph , start , goal ) :
  stack = ([[ start ]])
  visited = set()

  while stack :
    path = stack.pop()
    node = path [ -1]

    if node == goal :
      return path

    if node not in visited :
      visited.add ( node )

    for neighbor in graph[ node ]:
      new_path = list(path)
      new_path.append ( neighbor )
      stack.append(new_path)

  return None
print("BFS Path : ",bfs( graph , "A" , "F"))
```
